```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('12bb84ed-397e-435c-8d1b-8ce543108698.csv')
```

```python
df
```

```python
df['month_start'] = pd.to_datetime(df['month_start'])
```

```python
df
```

```python
df['month_start'].max()
```

```python
df['route_full_name'].unique()
```

```python
strip_district_routes = ['86 - LIBERTY', '87 - FRIENDSHIP', '88 - PENN', '54 - NORTH SIDE-OAKLAND-SOUTH SIDE', '91 - BUTLER STREET']
```

```python
df[df['route_full_name'].isin(strip_district_routes)].tail(10)
```

```python
df[df['route_full_name']=='86 - LIBERTY']
```

#### Liberty

```python
liberty = df[(df['route_full_name']=='86 - LIBERTY')]
monthly_liberty = liberty.groupby('month_start')['avg_riders'].sum().reset_index()
```

```python
monthly_liberty
```

```python
plt.figure(figsize=(20,6))
plt.plot(monthly_liberty['month_start'], monthly_liberty['avg_riders'], marker='o', linestyle='-')
```
